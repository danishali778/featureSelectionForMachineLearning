```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_validate

from feature_engine.selection import DropCorrelatedFeatures, SmartCorrelatedSelection
```

```python
data = pd.read_csv("gdrive/My Drive/Datasets/dataset_1.csv")
data
```

```python
X = np.array(data.iloc[:,:-1])
y = np.array(data.iloc[:,-1])
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
X_train.shape,X_test.shape
```

```python
sel = DropCorrelatedFeatures(threshold=0.8, method='pearson')
sel.fit(X_train)
```

```python
X_train_clean = sel.transform(X_train)
X_test_clean = sel.transform(X_test)
```

```python
print("Correlated features removed:", sel.features_to_drop_)
print("Original shape:", X_train.shape)
print("Cleaned shape:", X_train_clean.shape)
```

```python
print("Original shape:", X_test.shape)
print("Cleaned shape:", X_test_clean.shape)
```
