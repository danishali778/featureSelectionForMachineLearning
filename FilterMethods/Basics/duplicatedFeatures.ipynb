```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
# Manually Added two duplicated features
df['duplicate_1'] = df['sepal length (cm)']
df['duplicate_2'] = df['sepal width (cm)']
```

```python
import pandas as pd

def remove_duplicate_features(df: pd.DataFrame):
    """
    Remove duplicate columns from a DataFrame.

    Parameters:
    df (pd.DataFrame): Input DataFrame

    Returns:
    pd.DataFrame: DataFrame with duplicate columns removed
    list: List of removed column names
    """
    # Transpose, drop duplicate rows (which are original columns), and transpose back
    df_cleaned = df.T.drop_duplicates().T

    # Identify which columns were removed
    removed = list(set(df.columns) - set(df_cleaned.columns))

    return df_cleaned, removed
```

```python

cleaned_df, removed_columns = remove_duplicate_features(df)

print("‚úÖ Cleaned DataFrame:\n", cleaned_df.head())
print("\nüóëÔ∏è Removed Duplicate Features:", removed_columns)
```
