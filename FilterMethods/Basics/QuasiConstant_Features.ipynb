```python
class varianceThreshold:
  def __init__(self,threshold=0.0):
    self.threshold = threshold
    self.inputFeatures = None
    variances_ = None
    self.selectedFeatures = None
    self.constant_features = None

  def fit(self,X,y=None):
    nexamples,nfeatures = X.shape
    self.inputFeatures = nfeatures
    self.array = np.array(X)
    mean_array = np.mean(array,axis = 0)
    mean_array = mean_array.reshape(1,mean_array.shape[0])
    difference = array - mean_array
    square_diff = difference ** 2
    sum_of_sqDiff = np.sum(square_diff,axis = 0)
    variances = sum_of_sqDiff/nexamples
    constant_features = np.where(variances <= self.threshold)[0]
    self.selectedFeatures = np.where(variances > self.threshold)[0]
    self._variances = variances
    self.constant_features = constant_features

  def get_support(self,indices=None):
    mask = self._variances > self.threshold
    return mask

  def get_columnsNames(self,columns):
    return columns[self.selectedFeatures]



  def transform(self):
    X = self.array[:,self.selectedFeatures]
    return  X





```

```python
selector = varianceThreshold()
selector.fit(X)
```
